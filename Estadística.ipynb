## EJEMPLOS DE CLASE

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Ejemplo 1

**Generar números aleatorios**

```python
# Generar números a partir de una distribución exponencial, por ejemplo
numeros = np.random.exponential(scale = 1/12*60, size = 100000)

# Graficar
sns.histplot(numeros)
plt.show()
```

### Tomar muestras y calcular promedios

```python
# Generar muestras, por ejemplo, 200 listas de 50 elementos
chunk_size = 50
chunked_lists = [numeros[i:i + chunk_size] for i in range(0, len(numeros), chunk_size)]

# Comprobar que se hayan generado correctamente
len(chunked_lists), all(len(chunk) == 50 for chunk in chunked_lists)
```

```python
# Calcular promedios de las listas generadas
acum = []

for i in chunked_lists:
    acum.append(i.mean())
```

```python
acum
```

### Graficar promedios obtenidos

```python
# Generar histograma
sns.histplot(acum)
plt.show()
```
