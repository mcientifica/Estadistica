## EJERCICIO CORRELACIÃ“N

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
asignaturas = pd.read_excel("CorrelaciÃ³n.xlsx", sheet_name="Asignaturas")
monedas = pd.read_excel("CorrelaciÃ³n.xlsx", sheet_name="Monedas")
alturas = pd.read_excel("CorrelaciÃ³n.xlsx", sheet_name="Lectura vs Altura")
consumos = pd.read_excel("CorrelaciÃ³n.xlsx", sheet_name="ConsumoTV")
```

### Horas de Estudio vs Nota Obtenida

```python
# Mostrar primeros resultados
asignaturas.head()
```

```python
# Calcular correlaciÃ³n
np.corrcoef(x=asignaturas["HorasEstudio"], y=asignaturas["NotaObtenida"])
```

```python
# Graficar
sns.regplot(data=asignaturas, x="HorasEstudio", y="NotaObtenida", color='blue')
sns.scatterplot(data=asignaturas, x="HorasEstudio", y="NotaObtenida", color='red')
plt.show()
```
